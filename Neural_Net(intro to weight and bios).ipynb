#### numerical approach 

```python
def cost(b):
    return (b - 4) ** 2
```

```python
cost(4)
```

```python
def num_slope(b):
    h = 0.0001
    return (cost(b+h) - cost(b))/h
```

```python
num_slope(3)
```

```python
num_slope(5)
```

### Training loop

```python
b = 6
for i in range(100):
    b = b - .1 * slope(b)
    print(b)

# decreasing towards 4
```

```python
b = -2
for i in range(100):
    b = b - .1 * slope(b)
    print(b)

# increasing towards 4
```
